Importandos as dependências

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
```

Ajustando nossos dados

```python
receitas = [110000, 113000, 110000, 111000, 112000, 115000, 115000, 120000, 112000, 112000, 114000, 125000, 126000, 131000, 131000, 134000, 135000, 137000, 140000, 139000, 135000, 135000, 137000, 144000]
datas = pd.date_range(start="2023-01-01", periods=len(receitas), freq="ME")
df = pd.DataFrame({"Data": datas, "Receita": receitas})
```

Função para suavização exponencial simples

```python
def suavizacao_exponencial_simples(dados, alpha):
    resultado = [dados[0]] 
    
    for i in range(1, len(dados)):
        valor_suavizado = alpha * dados[i] + (1 - alpha) * resultado[i-1]
        resultado.append(valor_suavizado)
        
    return resultado
```

Aplicando a suavização com diferentes alphas

```python
alpha_07 = 0.7
alpha_05 = 0.5
alpha_03 = 0.3


valores_suavizados_07 = suavizacao_exponencial_simples(df['Receita'].tolist(), alpha=alpha_07)
valores_suavizados_05 = suavizacao_exponencial_simples(df['Receita'].tolist(), alpha=alpha_05)
valores_suavizados_03 = suavizacao_exponencial_simples(df['Receita'].tolist(), alpha=alpha_03)

df_resultado = pd.DataFrame({
    "Data": datas,
    "Receita Real": receitas,
    "Alpha 0.7": valores_suavizados_07,
    "Alpha 0.5": valores_suavizados_05,
    "Alpha 0.3": valores_suavizados_03

    })

df_resultado.head(30)
```

Calculando o RMSE dos modelos

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error

df_resultado['Previsao_0.7'] = df_resultado['Alpha 0.7'].shift(1)
df_resultado['Previsao_0.5'] = df_resultado['Alpha 0.5'].shift(1)
df_resultado['Previsao_0.3'] = df_resultado['Alpha 0.3'].shift(1)

rmse_07 = np.sqrt(mean_squared_error(df_resultado['Receita Real'][1:], df_resultado['Previsao_0.7'][1:]))
rmse_05 = np.sqrt(mean_squared_error(df_resultado['Receita Real'][1:], df_resultado['Previsao_0.5'][1:]))
rmse_03 = np.sqrt(mean_squared_error(df_resultado['Receita Real'][1:], df_resultado['Previsao_0.3'][1:]))

print("RMSE Preditivo (Cálculo Correto)")
print(f'RMSE (Alpha 0.7): R$ {rmse_07:,.2f}')
print(f'RMSE (Alpha 0.5): R$ {rmse_05:,.2f}')
print(f'RMSE (Alpha 0.3): R$ {rmse_03:,.2f}')
```

Fazendo a previsão de receita para Janeiro de 2025

```python
previsao_jan_2025_alpha_07 = valores_suavizados_07[-1]
previsao_jan_2025_alpha_05 = valores_suavizados_05[-1]
previsao_jan_2025_alpha_03 = valores_suavizados_03[-1]

print("Previsão de Receita para Janeiro de 2025")
print(f"Usando Alpha = 0.7: R$ {previsao_jan_2025_alpha_07:,.2f}")
print(f"Usando Alpha = 0.5: R$ {previsao_jan_2025_alpha_05:,.2f}")
print(f"Usando Alpha = 0.3: R$ {previsao_jan_2025_alpha_03:,.2f}")
```
